````python
def jaccard_similarity(str1, str2):
    # Tokenize both strings and convert to sets
    set1 = set(str1.split())
    set2 = set(str2.split())
    # Calculate intersection and union
    intersection = set1.intersection(set2)
    union = set1.union(set2)
    # Calculate Jaccard score
    return len(intersection) / len(union) if union else 0

# Apply Jaccard similarity calculation for each row
data['jaccard_score'] = data.apply(lambda row: jaccard_similarity(row['cleaned_text'], row['boolean_query']), axis=1)

# Display the results
print("\nGenerated Boolean Queries and Jaccard Scores:")
print(data[['input', 'boolean_query', 'jaccard_score']].head())
````
